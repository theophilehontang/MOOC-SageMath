```python
def fermat(n):
    for j in [2..n-1]:
        if (j^(n-1)).mod(n)!=1:
            return(j)
    return()
```

```python
def chinois(u,v):
    n=1
    e=0
    for j in range(len(u)):
        n=n*u[j]
    for b in range(len(u)):
        o=n/u[b]
        e=o*xgcd(o,u[b])[1]*v[b]+e
    return(mod(e,n))
        
```
